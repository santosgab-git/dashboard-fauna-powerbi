```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 1500
fechas = pd.date_range("2023-01-01", "2023-12-31", periods=n)

areas = ["Área 1", "Área 2", "Área 3"]
actividades = ["rocería", "obra civil", "tránsito maquinaria", "inspección"]

especies_data = [
    ("Boa constrictor", "Boa", "Reptilia"),
    ("Crotalus durissus", "Cascabel", "Reptilia"),
    ("Iguana iguana", "Iguana", "Reptilia"),
    ("Chelonoidis carbonaria", "Morrocoy", "Reptilia"),
    
    ("Didelphis marsupialis", "Zarigüeya", "Mammalia"),
    ("Procyon cancrivorus", "Mapache", "Mammalia"),
    ("Dasypus novemcinctus", "Armadillo", "Mammalia"),
    ("Tamandua mexicana", "Oso hormiguero", "Mammalia"),
    
    ("Coragyps atratus", "Chulo", "Aves"),
    ("Milvago chimachima", "Caracara", "Aves"),
    ("Columbina talpacoti", "Torcaza", "Aves"),
    ("Ardea alba", "Garza blanca", "Aves"),
    
    ("Rhinella marina", "Sapo común", "Amphibia"),
    ("Leptodactylus fragilis", "Rana", "Amphibia"),
    ("Hypsiboas crepitans", "Rana arborícola", "Amphibia"),
    ("Scinax ruber", "Ranita", "Amphibia"),
    
    ("Anolis auratus", "Camaleoncito", "Reptilia"),
    ("Erythrolamprus bizona", "Culebra", "Reptilia"),
    ("Zygodontomys brevicauda", "Ratón de caña", "Mammalia"),
    ("Turdus ignobilis", "Mirla", "Aves")
]

data = []

for i in range(n):
    fecha = fechas[i]
    
    especie, nombre, clase = especies_data[np.random.randint(len(especies_data))]
    
    actividad = np.random.choice(actividades)

    clase_probs = {
    "Reptilia": 0.31,
    "Amphibia": 0.04,
    "Aves": 0.60,
    "Mammalia": 0.05
}

    clase = np.random.choice(list(clase_probs.keys()), p=list(clase_probs.values()))

    especies_filtradas = [e for e in especies_data if e[2] == clase]
    especie, nombre, clase = especies_filtradas[np.random.randint(len(especies_filtradas))]
    
    # Manejo según clase
    if clase == "Reptilia":
        manejo = np.random.choice(["rescate", "ahuyentamiento"], p=[0.25, 0.75])

    elif clase == "Amphibia":
        manejo = np.random.choice(["rescate", "ahuyentamiento"], p=[0.99, 0.01])

    elif clase == "Aves":
        manejo = np.random.choice(["rescate", "ahuyentamiento"], p=[0.01, 0.99])

    elif clase == "Mammalia":
        manejo = np.random.choice(["rescate", "ahuyentamiento"], p=[0.25, 0.75])
    
    # Tendencia por mes
    mes = fecha.month
    if mes <= 3:
        abundancia = np.random.randint(2, 5)
    elif mes <= 6:
        abundancia = np.random.randint(1, 4)
    else:
        abundancia = np.random.randint(1, 3)
    
    # Mortalidad por actividad

    if manejo == "rescate":
        if actividad in ["rocería", "tránsito maquinaria"]:
            estado = np.random.choice(["vivo", "muerto"], p=[0.92, 0.08])
        else:
            estado = np.random.choice(["vivo", "muerto"], p=[0.98, 0.02])
    else:
        estado = "vivo"
    
    if manejo == "rescate" and estado == "vivo":
        reubicado = "si"
    elif manejo == "rescate":
        reubicado = "no"
    else:
        reubicado = None
    
    data.append([
        i+1,
        fecha.strftime("%Y-%m-%d"),
        "Proyecto Solar Alpha",
        np.random.choice(areas),
        actividad,
        manejo,
        clase,
        especie,
        nombre,
        abundancia,
        estado,
        reubicado
    ])

df = pd.DataFrame(data, columns=[
    "id_registro","fecha","proyecto","area","actividad",
    "manejo","clase","especie","nombre_comun",
    "abundancia","estado","reubicado"
])

df.columns = [col.capitalize() for col in df.columns]

df.to_csv("Registros.csv", index=False)
```
